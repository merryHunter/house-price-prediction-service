## Conclusion:

X5 and X6 are correlated, but wee see that houses in train set is only from same region. By including these features we can have problems if houses come from new regions.

```python
df = create_features(df)
```

```python
X_train = df[['X2 house age', 'X3 distance to the nearest MRT station' , 'X4 number of convenience stores']]
y_train = df[['Y house price of unit area']]
```

```python
from sklearn.model_selection import train_test_split
X_train, X_val, y_train, y_val = train_test_split(X_train, y_train, test_size=0.2, random_state=42)
```

```python
from sklearn.pipeline import Pipeline, make_pipeline
from sklearn.preprocessing import StandardScaler
from sklearn.linear_model import Ridge
```

```python
ml_pipeline = make_pipeline(StandardScaler(), Ridge())
```

```python
ml_pipeline.fit(X_train, y_train)
```

```python
X_pred = ml_pipeline.predict(X_val)
X_pred[0]
```

```python
# metrics
```

```python
from sklearn.metrics import r2_score
```

```python
r2_score(X_pred, y_val)
```
